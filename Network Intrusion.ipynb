```python
import pandas as pd

# Assuming 'df' is your DataFrame containing numerical columns
# Replace this with your actual DataFrame
numerical_cols = ['duration', 'src_bytes', 'dst_bytes', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate']

# Define a function to detect outliers using the IQR method
def find_outliers_iqr(df, numerical_cols, threshold=1.5):
    outlier_count = {}
    for col in numerical_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (threshold * iqr)
        upper_bound = q3 + (threshold * iqr)
        outliers_col = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count[col] = len(outliers_col)
    return outlier_count

# Detect outliers in numerical columns using the IQR method
outlier_count_iqr = find_outliers_iqr(df, numerical_cols)

# Display the number of outliers for each column
print("Number of outliers detected using IQR method:")
for col, count in outlier_count_iqr.items():
    print(f"{col}: {count}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to remove outliers using modified IQR method
def remove_outliers_modified(df, var):
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 3 * IQR  # Larger multiplier for lower bound
    ub = Q3 + 3 * IQR  # Larger multiplier for upper bound
    df_cleaned = df[(df[var] >= lb) & (df[var] <= ub)]
    return df_cleaned

# Define the variables for which you want to remove outliers
variables = ['duration', 'src_bytes', 'dst_bytes', 'count', 'srv_count', 
             'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 
             'diff_srv_rate', 'srv_diff_host_rate']

# Remove outliers
for var in variables:
    df = remove_outliers_modified(df, var)

# Create box plots after removing outliers
plt.figure(figsize=(20, 15))
for i, var in enumerate(variables):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(y=df[var], color='skyblue')
    plt.title(f"After Removal: {var}")

# Adjust spacing between subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

plt.show()
```
