# Data Preprocessing

```python
# Combining X and y into single dataframe
import pandas as pd

df = pd.concat([pd.DataFrame(X), pd.DataFrame(y)], axis=1)
df
```

```python
# Number of independent and dependent variables
indep_var = df.columns[:-1]
indep_var_count = len(indep_var)
print('Number of independent variables: ', indep_var_count)

dep_var = df.columns[-1:]
dep_var_count = len(dep_var)
print('Number of dependent variables: ', dep_var_count)
```

```python
# Number of records
num_records = len(df)
print("Number of records:", num_records)
```

```python
df.dtypes
```

```python
binary_columns = []
nominal_columns = []
categorical_columns = []
textual_columns = []
numerical_columns = []

for column, dtype in df.dtypes.items():
    if dtype == 'bool':
        binary_columns.append(column)
    elif dtype == 'object':
        if len(df[column].unique()) == 2:
            binary_columns.append(column)
        else:
            nominal_columns.append(column)
    elif dtype == 'category':
        categorical_columns.append(column)
    elif dtype == 'float64' or dtype == 'int64':
        numerical_columns.append(column)
    elif dtype == 'string':
        textual_columns.append(column)

print("Binary columns:", binary_columns)
print("\nNominal columns:", nominal_columns)
print("\nCategorical columns:", categorical_columns)
print("\nTextual columns:", textual_columns)
print("\nNumerical columns:", numerical_columns)
```

```python
# Encoding the categorical variables
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

label_encoder = LabelEncoder()

for column in df_encoded.columns:
    if df_encoded[column].dtype == 'object':
        df_encoded[column] = label_encoder.fit_transform(df_encoded[column])
```

```python
# Splitting data into features and target
X = df_encoded.drop(columns=['Attack_type'])
y = df_encoded['Attack_type']
```

```python
# Scaling the numerical features
from sklearn.preprocessing import StandardScaler

numeric_columns = X.select_dtypes(include=['int64', 'float64'])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric_columns)
X_scaled_df = pd.DataFrame(X_scaled, columns=numeric_columns.columns)

categorical_columns = X.select_dtypes(exclude=['int64', 'float64'])

X_scaled_df.index = categorical_columns.index = y.index

# Combining numerical and categorical columns of X
combined_X = pd.concat([X_scaled_df, categorical_columns], axis=1)

# Combining X and y as a dataframe
combined_df = pd.concat([combined_X, y], axis=1)
```

```python
numerical_columns = combined_df.select_dtypes(include=['int64', 'float64']).columns

def find_outliers_iqr(df, numerical_columns, threshold=1.5):
    outlier_count = {}
    for col in numerical_columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (threshold * iqr)
        upper_bound = q3 + (threshold * iqr)
        outliers_col = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count[col] = len(outliers_col)
    return outlier_count

outlier_count_iqr = find_outliers_iqr(combined_df, numerical_columns)
total_values_per_column = combined_df[numerical_columns].count()
outliers_proportion = {col: outlier_count_iqr[col] / total_values_per_column[col] * 100 for col in numerical_columns}

print("Number of outliers detected using IQR method:")
for col, count in outlier_count_iqr.items():
    print(f"{col}: {count}")
print("\nProportion of outliers in each numerical column (%):")
for col, proportion in outliers_proportion.items():
    print(f"{col}: {proportion:.2f}%")
```
