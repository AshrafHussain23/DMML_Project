# Comparison of Metrics of All Models

```python
import numpy as np
import matplotlib.pyplot as plt

models = ['Random Forest', 'XGBoost', 'SVM', 'LIBSVM']
accuracy_values = [0.9754856614246068, 0.973172987974098, 0.9518963922294172, 0.9315448658649399]
f1_score_values = [0.9754200001781264, 0.9731205514139859, 0.9512533491117061, 0.9300660389742388]

bar_width = 0.35
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(10, 6))

plt.bar(r1, accuracy_values, color='skyblue', width=bar_width, edgecolor='grey', label='Accuracy')
plt.bar(r2, f1_score_values, color='lightgreen', width=bar_width, edgecolor='grey', label='F1 Score')

for i in range(len(models)):
    plt.text(x=r1[i]-0.02, y=accuracy_values[i]+0.01, s=str(round(accuracy_values[i], 3)), fontsize=10)
    plt.text(x=r2[i]-0.02, y=f1_score_values[i]+0.01, s=str(round(f1_score_values[i], 3)), fontsize=10)

plt.xlabel('Models', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(models))], models, rotation=45)

plt.legend(loc='lower right', bbox_to_anchor=(1, 0))

plt.title('Comparison of Accuracy and F1 Score for Different Models of Network')
plt.tight_layout()
plt.show()
```
